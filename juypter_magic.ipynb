```python
from IPython.core.magic import register_line_magic
import inspect

@register_line_magic
def p(line):
    """Usage: %p ClassName"""
    obj = eval(line)
    try:
        sig = inspect.signature(obj)
    except:
        print(f"âŒ Cannot inspect")
        return
    
    print(f"\nðŸ“¦ {line}\n" + "="*40)
    
    print("\nðŸ”´ REQUIRED:")
    for n, param in sig.parameters.items():
        if n not in ('self','cls','args','kwargs'):
            if param.default == inspect.Parameter.empty:
                print(f"   â€¢ {n}")
    
    print("\nðŸŸ¢ OPTIONAL:")
    for n, param in sig.parameters.items():
        if n not in ('self','cls','args','kwargs'):
            if param.default != inspect.Parameter.empty:
                print(f"   â€¢ {n} = {repr(param.default)}")
```
